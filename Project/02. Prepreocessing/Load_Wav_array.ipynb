```python
import librosa
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import pandas as pd
```

```python
filename = np.array(pd.read_csv('../AI_Project_Source_Code/data/SC_file_name.csv'))
answer = np.array(pd.read_csv('../AI_Project_Source_Code/data/SC_answer.csv'))
for step in range(186):
    audio_path = '../AI_Project_Source_Code/data/'+filename[step]
    if (step>100) & (step<180):
        y, sr = librosa.load(audio_path[0]) 
        D = np.abs(librosa.stft(y))
        librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='mel', x_axis='time')
        print(answer[step])
        plt.title(audio_path[0])
        plt.show()
 
```
