```python
import librosa
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import pandas as pd
```

```python
train_input = []
train_label = []


filename = np.array(pd.read_csv('/data/SC_file_name.csv'))
answer = np.array(pd.read_csv('../AI_Project_Source_Code/data/SC_answer.csv'))
path = './MFCC_IMAGE/'
for step in range(190):
    audio_path = '../AI_Project_Source_Code/data/'+filename[step]
    signal, sr = librosa.load(audio_path[0], sr=44100 , res_type = 'kaiser_best')
    mfcc = librosa.feature.mfcc(y=signal, sr=sr)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    
    fig = plt.figure(figsize=(10,10))
    librosa.display.specshow(mfcc_delta2)
    if answer[step] == 0 :
        plt.savefig(path+'train_0/'+''.join(filename[step])+'.png')
    else:
        plt.savefig(path+'train_1/'+''.join(filename[step])+'.png')
        

filename = np.array(pd.read_csv('data/SC_file_name.csv'))
answer = np.array(pd.read_csv('data/SC_answer.csv'))
for step in range(0,36):
    audio_path = 'data/wav/'+filename[step]
    y, sr = librosa.load(audio_path[0])
    D = np.abs(librosa.stft(y))
    librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max))
    print(answer[step+1])
    #plt.title(audio_path[0])
    plt.show()
 
```
